```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
data='../data/'
```

```python
ipsc_trans = pd.read_csv(data+'ips_tpm.csv')
ipsc_trans.head(10)
```

```python
ipsc_trans.shape
```

```python
nonipsc_trans = pd.read_csv(data+'all_tpm.csv.gz', compression='gzip')
nonipsc_trans.head()
```

```python
# get most expressed tfs
top_genes = pd.read_csv(data+'top20tf_ens.csv')
```

```python
top_genes.head()
```

```python
top_tnames = top_genes['Gene name'].tolist()
top_tnames
```

```python
nonipsc_20 = nonipsc_trans[top_tnames]
ipsc_20 = ipsc_trans[top_tnames]
```

```python
nonipsc_20.head()
```

```python
nonipsc_20['label'] = 0
ipsc_20['label'] = 1
data_20 = ipsc_20.append(nonipsc_20, ignore_index=True)
data_shuffled = data_20.sample(frac=1)
```

```python
data_shuffled.shape
```

```python
train_test_cutoff = int(data_shuffled.shape[0]/5)
```

```python
train_test_cutoff
```

```python
data_shuffled.head()
```

```python
test_set = data_shuffled.iloc[:train_test_cutoff,:]
train_set = data_shuffled.iloc[train_test_cutoff:,:]
```

```python
X_train = train_set.iloc[:,:-1]
y_train = train_set.iloc[:,-1]
```

```python
X_test = test_set.iloc[:,:-1]
y_test = test_set.iloc[:,-1]
```

```python
test_set.head()
```

```python
y_train.value_counts()
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
y_probs = clf.predict_proba(X_test)
```
